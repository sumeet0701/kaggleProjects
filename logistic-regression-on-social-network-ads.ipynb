## Logistic Regression

```python
x = ads_data[['Gender','Age','EstimatedSalary']]
y = ads_data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler() 
X_scaled = ss.fit_transform(x) 
X_scaled_train, X_scaled_test, y_scaled_train, y_scaled_test = train_test_split(X_scaled, y,train_size=0.33,random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
```

```python
lr.fit(X_scaled_train, y_scaled_train)
```

```python
prediction1 = lr.predict(X_scaled_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
acc = accuracy_score(y_test, prediction)
print(f'Accuracy: {acc}')

acc1 = accuracy_score(y_scaled_test, prediction1)
print(f'Accuracy1: {acc1}')
```

```python
print(classification_report(y_test,prediction))
```

```python
print(classification_report(y_scaled_test,prediction1))
```
